### 문제 
- 크기 
$N\times N$의 격자가 있다. 정확히 한 행만 선택하여 그 행에서 원하는 만큼의 칸을 색칠할 수 있다.

양쪽 대각선 모두에 색칠된 칸이 존재하도록 만들고자 한다. 
$i$행 
$j$열의 칸을 
$(i, j)$로 나타낼 때, 양쪽 대각선은 
$\{(1, 1), (2, 2), \ldots, (N, N)\}$과 
$\{(1, N), (2, N-1), \ldots, (N, 1)\}$을 의미한다.

칠해야 하는 칸의 최소 개수를 구하여라.

### 입력 
- 첫째 줄에 양의 정수 
$N(2\leq N\leq 10)$이 주어진다.

### 출력 
- 첫째 줄에 조건을 만족하기 위해 칠해야 하는 칸의 최소 개수를 출력한다.



```python
N = int(input())
print(1 if N % 2 == 1 else 2)
```
